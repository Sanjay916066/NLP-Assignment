```python
import pandas as pd

documents = [
    {
        "title": "Galaxy Warriors",
        "description": "A space adventure where robots help humans fight alien enemies."
    },
    {
        "title": "Healthy Living Guide",
        "description": "Tips for healthy food, exercise, and balanced nutrition."
    },
    {
        "title": "Robot Future",
        "description": "Advanced robots transform life in a futuristic smart city."
    },
    {
        "title": "Cooking Masterclass",
        "description": "Learn delicious healthy recipes and cooking techniques."
    },
    {
        "title": "Deep Space Journey",
        "description": "Astronauts explore distant planets during an exciting mission."
    },
    {
        "title": "Fitness Revolution",
        "description": "Daily workouts and healthy eating habits for better fitness."
    },
    {
        "title": "AI and Machines",
        "description": "Artificial intelligence and robots change modern industries."
    },
    {
        "title": "Ocean Mystery",
        "description": "Divers discover hidden treasures under the sea."
    },
    {
        "title": "Interstellar Rescue",
        "description": "A brave team travels through space to rescue missing scientists."
    },
    {
        "title": "Vegan Food Secrets",
        "description": "Healthy plant-based meals with nutrition advice and recipes."
    }
]

df = pd.DataFrame(documents)

print(df)
```

```python
import string
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Download required files
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

# Stopwords and Lemmatizer
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# Preprocessing function
def preprocess_text(text):

    # Convert to lowercase
    text = text.lower()

    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Split into words
    words = text.split()

    # Remove stopwords and lemmatize
    words = [
        lemmatizer.lemmatize(word)
        for word in words
        if word not in stop_words
    ]

    # Join words back
    return " ".join(words)

# Combine title and description
df["combined_text"] = df["title"] + " " + df["description"]

# Apply preprocessing
df["processed_text"] = df["combined_text"].apply(preprocess_text)

# Show processed text
print(df[["title", "processed_text"]])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create TF-IDF Vectorizer
vectorizer = TfidfVectorizer()

# Convert processed text into TF-IDF matrix
tfidf_matrix = vectorizer.fit_transform(df["processed_text"])

# Show shape of matrix
print("TF-IDF Matrix Shape:", tfidf_matrix.shape)
```

```python
# User query
user_query = "space adventure with robots"

# Preprocess query
processed_query = preprocess_text(user_query)

print("Processed Query:", processed_query)

# Convert query into TF-IDF vector
query_vector = vectorizer.transform([processed_query])

print("Query Vector Shape:", query_vector.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate cosine similarity
similarity_scores = cosine_similarity(query_vector, tfidf_matrix)

# Display similarity scores
print(similarity_scores)
```

```python
# Convert similarity scores into a 1D list
scores = similarity_scores[0]

# Get indices of top 3 similar documents
top_indices = scores.argsort()[-3:][::-1]

print("Top 3 Recommended Documents:\n")

# Display results
for idx in top_indices:
    print("Title:", df.iloc[idx]["title"])
    print("Description:", df.iloc[idx]["description"])
    print("Similarity Score:", round(scores[idx], 4))
    print("-" * 50)
```
